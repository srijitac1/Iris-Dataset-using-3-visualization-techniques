```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the built-in dataset
df = sns.load_dataset('iris')

# Display basic structure
print("Dataset Shape:", df.shape)
print("\nFirst 5 Rows:")
print(df.head())
print("\nMissing Values:")
print(df.isnull().sum())
```

**Visualization 1 - Pair Plot (Scatter Plot Matrix) :** A Pair Plot visualizes pairwise relationships across an entire dataset. It creates a matrix of scatter plots for every combiation of features, with histograms along the diagonal to show univariate distributions.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = sns.load_dataset('iris')

# Generate the Pair Plot colored by species
sns.pairplot(df, hue='species', palette='Set2')
plt.suptitle("Pair Plot of Iris Features", y=1.02)
plt.show()
```

**Technique 2 - Box Plot (Distribution & Outlier Analysis) :** Box Plots are exceptional for comparing the distribution, median, variance, and identifying outliers across different categories (species) for a specific feature.

```python
plt.figure(figsize=(10, 6))
# Distribution of Petal Length for each species
sns.boxplot(x='species', y='petal_length', data=df, hue='species', palette='Pastel1', legend = False)
plt.title('Petal Length Distribution by Species')
plt.xlabel('Species')
plt.ylabel('Petal Length (cm)')
plt.show()
```

**Technique 3 - Correlation Heatmap :** A heatmap represents a correlation matrix graphically using colors. It allows data miners to instantly detect multi - collinearity (highly correlated features), which is crucial for feature selection in machine learning tasks.

```python
plt.figure(figsize=(8,6))
# Calculate correlation matrix on numeric columns
corr_matrix = df.drop(columns = 'species').corr()

# Generate the heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap of Iris Numeric Features')
plt.show()
```
